```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import  LinearRegression
```

```
#. Predicting House Prices
#●	Input: Size of the house in square feet, number of bedrooms.
#●	Output: Predicted price of the house

size = np.array([800,1000,1200,1400,1600]).reshape(-1,1)
prices =np.array([15000,20000,25000,28000,30000])
```

```
model = LinearRegression()
```

```
model
```

```
model.fit(size,prices)
```

```
prediction_value = np.array([[1800]])
```

```
prediction_price = model.predict(prediction_value)
```

```
prediction_price
```
